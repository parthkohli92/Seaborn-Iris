# Seaborn

1. Displot
2. Joinplot
3. Pairplot

```python
import seaborn as sns
```

```python
a = sns.load_dataset("tips")
```

```python
a
```

```python
a.head()
```

## Correlation with heatmap

A corr heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. it is very important in feature selection

```python
a.corr()
```

```python
a.dtypes
```

```python
sns.heatmap(a.corr())
```

## Joint Plot- Univariabe analysis

```python
sns.jointplot(x='tip',y='total_bill',data=a,kind='reg')
```

## Pair Plot

```python
sns.pairplot(a, kind='reg')
```

```python
sns.pairplot(a, hue='day', kind='hist')
```

## Dist Plot

```python
sns.displot(a['tip'], kind='kde')
```

```python
sns.displot(a['tip'],kde=True,bins=10)
```

# Categorical Plots

1.Boxplot
2.Violinplot
3.Joinplot
4.Barplot

```python
a.head()
```

```python
sns.countplot('day', data=a)
```

```python
## count plot

sns.countplot(y='day',data=a)
```

```python
## Bar plot 

sns.barplot(x='smoker', y='total_bill',data=a)
```

```python
## Box plot 

sns.boxplot('tip','day',data=a)
```

```python
sns.boxplot(x='tip', y='sex',hue='smoker',data=a)
```

```python
## Violin plot

sns.violinplot(x='tip',y='sex',hue='smoker',data=a)
```

```python
## box plot 

sns.jointplot(x='tip',y='total_bill',hue='smoker',data=a)
```

# Using Iris Data set for Seaborn

```python
##Loading data

iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
## Joint PLot

sns.jointplot(x='petal_width',y='petal_length',data=iris, kind='reg')
```

```python
## Displot

sns.displot(iris, kind='hist')
```

```python
## Pair plot 

sns.pairplot(hue='species',data=iris)
```

```python
## Count plot

sns.countplot(x='sepal_length', orient="v",data=iris, hue='species')
```

```python
## Violinplot

sns.violinplot(x='sepal_length',data=iris, y="sepal_width" , hue='species')
```

```python
## boxplot

sns.boxplot(x='sepal_length',data=iris, y="sepal_width")
```
